```python
import pymongo
import json
client=pymongo.MongoClient('mongodb://localhost:27017/')
mydb1=client['Students_database']
collec1=mydb1['students_record']
records=[]
with open('students.json') as f:
    for line in f:
        dic1=json.loads(line)
        records.append(dic1)

collec1.insert_many(records)
```

```python
for i in collec1.find({}):
    print(i)
```

```python

#1 find the student with highest mark
high=collec1.aggregate([{"$unwind":"$scores"},
    {'$group':{"_id":"$name",'Total':{'$sum':'$scores.score'}}},
                        {'$sort':{'Total':-1}},
                        {'$limit':1}
])
for i in high:
    print(i)
```

```python
averageofexam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{'_id':'ae','averageofexam':{'$avg':'$scores.score'}}}
                        
                        
])
for i in averageofexam:
    print(i)
    
averageofquiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{'_id':'aq','averageofquiz':{'$avg':'$scores.score'}}}
                        
                        
])
for i in averageofquiz:
    print(i)
averageofhomework=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{'_id':'ah','averageofhomework':{'$avg':'$scores.score'}}}
                        
                        
])
for i in averageofhomework:
    print(i)    
```

```python
average_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Average':{'$avg':'$scores.score'}}},
    {'$match':{'Average':{'$gte':40,"$lt":48.67}}}
                        
                        
])
for i in average_exam:
                print(i)
```

```python
pass_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Average':{'$avg':'$scores.score'}}},
    {'$project':{'name':1,'Average':1,'passorfail':{"$cond":[{'$lt':["$Average",40]},'Fail','Pass']}}},
    {'$sort':{'_id':1}}
                        
                        
])
for i in pass_exam:
                print(i)
       
```

```python
pass_quiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{"_id":"$name",'Average':{'$avg':'$scores.score'}}},
    {'$project':{'name':1,'Average':1,'passorfail':{"$cond":[{'$lt':["$Average",40]},'Fail','Pass']}}},
    {'$sort':{'_id':1}}
    
                        
                        
])
for i in pass_quiz:
                print(i)
```

```python
pass_homework=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{"_id":"$name",'Average':{'$avg':'$scores.score'}}},
    {'$project':{'name':1,'Average':1,'passorfail':{"$cond":[{'$lt':["$Average",40]},'Fail','Pass']}}},
    {'$sort':{'_id':1}}
    
                        
                        
])
for i in pass_homework:
                print(i)
```

```python
Total_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Total_exam':{'$sum':'$scores.score'},'Average_exam':{'$avg':'$scores.score'}}},
    {'$out':'TotalandAverage'}
    
                        
                        
])
Total_quiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{"_id":"$name",'Total_quiz':{'$sum':'$scores.score'},'Average_quiz':{'$avg':'$scores.score'}}},
    {'$merge':{'into':'TotalandAverage','on':'_id'}}
    
                        
                        
])
Total_homework=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{"_id":"$name",'Total_homework':{'$sum':'$scores.score'},'Average_homework':{'$avg':'$scores.score'}}},
    {'$merge':{'into':'TotalandAverage','on':'_id'}}
    
                        
                        
])
collec2=mydb1['TotalandAverage']
for i in collec2.find():
    print(i)
```

```python
Belowaverage_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Average_exam':{'$avg':'$scores.score'}}},
    {'$match':{'Average_exam':{'$gte':40,"$lt":48.67}}},
    {'$out':'BelowAverage'}
                        
                        
])
Belowaverage_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{"_id":"$name",'Average_quiz':{'$avg':'$scores.score'}}},
    {'$match':{'Average_quiz':{'$gte':40,"$lt":48.99}}},
    {'$merge':{'into':'BelowAverage','on':'_id'}}
                        
                        
])
Belowaverage_quiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{"_id":"$name",'Average_homework':{'$avg':'$scores.score'}}},
    {'$match':{'Average_homework':{'$gte':40,"$lt":67.81}}},
    {'$merge':{'into':'BelowAverage','on':'_id'}}
                        
                        
])
collec3=mydb1['BelowAverage']
for i in collec3.find():
    print(i)
```

```python
Fail_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Average_exam':{'$avg':'$scores.score'}}},
    {'$match':{'Average_exam':{"$lt":40}}},
    {'$out':'Failedstudents'}
                        
                        
])
Failed_quiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{"_id":"$name",'Average_quiz':{'$avg':'$scores.score'}}},
    {'$match':{'Average_quiz':{"$lt":40}}},
    {'$merge':{'into':'Failedstudents','on':'_id'}}
                        
                        
])
Failed_homework=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{"_id":"$name",'Average_homework':{'$avg':'$scores.score'}}},
    {'$match':{'Average_homework':{"$lt":40}}},
    {'$merge':{'into':'Failedstudents','on':'_id'}}
                        
                        
])
collec4=mydb1['Failedstudents']
for i in collec4.find():
    print(i)
```

```python
Passed_exam=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'exam'}},
    {'$group':{"_id":"$name",'Average_exam':{'$avg':'$scores.score'}}},
    {'$match':{'Average_exam':{"$gte":40}}},
    {'$out':'Passedstudents'}
                        
                        
])
Passed_quiz=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'quiz'}},
    {'$group':{"_id":"$name",'Average_quiz':{'$avg':'$scores.score'}}},
    {'$match':{'Average_quiz':{"$gte":40}}},
    {'$merge':{'into':'Passedstudents','on':'_id'}}
                        
                        
])
Passed_homework=collec1.aggregate([
    {"$unwind":"$scores"},
    {'$match':{"scores.type":'homework'}},
    {'$group':{"_id":"$name",'Average_homework':{'$avg':'$scores.score'}}},
    {'$match':{'Average_homework':{"$gte":40}}},
    {'$merge':{'into':'Passedstudents','on':'_id'}}
                        
                        
])
collec5=mydb1['Passedstudents']
for i in collec5.find():
    print(i)
```
